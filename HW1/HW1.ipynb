```python
import pandas as pd
import altair as alt
```

```python
alt.renderers.enable('default')
```

```python
df = pd.read_csv("GLB.Ts_dSST.csv", header=1)
```

```python
df
```

```python
df = pd.melt(df, id_vars=['Year'], value_vars=['J-D'])
```

```python
df
```

```python
x_axis = [el for el in range(df['Year'].min(), df['Year'].max(), 10)]
```

```python
plot = alt.Chart(
    df,
    title=['Annual Global Surface Temperature,', 'Relative to Late 19th Century Average']).mark_point().encode(
    alt.X(
        field='Year', 
        title=None,
        type='ordinal', 
        axis=alt.Axis(values=x_axis, grid=True, labelAngle=0)
        ),
    alt.Y(
        field='value', 
        title=None,
        type='quantitative', 
        axis=alt.Axis(labelExpr="(datum.value > 0 ? '+' + datum.value + '\u00b0' "
                      ": (datum.value == -0.6) ? null: datum.value + '\u00b0')", tickCount=10)
    )
)
```

```python
axis_text = alt.Chart(pd.DataFrame([{'x': 2010, 'y': 0}])).mark_text(
    text=['HOTTER THAN THE', '1880-1899 AVERAGE', '', 'COLDER'],
    dy = -25
).encode(
    x='x:O',
    y='y:Q'
)
```

```python
ox_axis = alt.Chart(pd.DataFrame({'y': [0]})).mark_rule().encode(y='y')
```

```python
right_side_labels = alt.Chart(
    df,
    ).mark_text(dx=20).transform_filter(
    alt.FieldOneOfPredicate(field='Year', oneOf=[1944, 2014, 2015, 2017])
).encode(
    alt.X(
        field='Year', 
        title=None,
        type='ordinal', 
        axis=alt.Axis(values=x_axis, grid=True, labelAngle=0)
        ),
    alt.Y(
        field='value', 
        title=None,
        type='quantitative', 
        axis=alt.Axis(labelExpr="(datum.value > 0 ? '+' + datum.value + '\u00b0' "
                      ": (datum.value == -0.6) ? null: datum.value + '\u00b0')", tickCount=10)
    ),
    alt.Text('Year:O')
)
```

```python
left_side_labels = alt.Chart(
    df,
    ).mark_text(dx=-20).transform_filter(
    alt.FieldOneOfPredicate(field='Year', oneOf=[1904, 1998])
).encode(
    alt.X(
        field='Year', 
        title=None,
        type='ordinal', 
        axis=alt.Axis(values=x_axis, grid=True, labelAngle=0)
        ),
    alt.Y(
        field='value', 
        title=None,
        type='quantitative', 
        axis=alt.Axis(labelExpr="(datum.value > 0 ? '+' + datum.value + '\u00b0' "
                      ": (datum.value == -0.6) ? null: datum.value + '\u00b0')", tickCount=10)
    ),
    alt.Text('Year:O')
)
```

```python
bold_labels = alt.Chart(
    df,
    ).mark_text(dx=20, fontWeight="bold").transform_filter(
    alt.FieldOneOfPredicate(field='Year', oneOf=[2016])
).encode(
    alt.X(
        field='Year', 
        title=None,
        type='ordinal', 
        axis=alt.Axis(values=x_axis, grid=True, labelAngle=0)
        ),
    alt.Y(
        field='value', 
        title=None,
        type='quantitative', 
        axis=alt.Axis(labelExpr="(datum.value > 0 ? '+' + datum.value + '\u00b0' "
                      ": (datum.value == -0.6) ? null: datum.value + '\u00b0')", tickCount=10)
    ),
    alt.Text('Year:O')
)
```

```python
alt.layer(plot, ox_axis, axis_text, right_side_labels, left_side_labels, bold_labels).properties(
        width=720, 
        height=400,
    ).configure_point(
        color='gray',
        fill='#E55B13',
        strokeWidth=0.8
    ).configure_title(
        anchor='start',
        dx = 50,
        dy = 50,
        fontWeight = 800
    ).configure_axis(
        labelColor='gray',
        ticks = False,
        domain=False,
    ).configure_view(
    strokeOpacity=0
    )
```
